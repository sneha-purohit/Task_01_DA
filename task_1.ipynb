```python
import pandas as pd
```

```python
# Step 1: Load the dataset
df = pd.read_csv(r"C:\Users\sneha\Downloads\Netflix.csv")
```

```python
# Step 2: View basic info
print(df.info())
print(df.head())
```

```python
# Step 3: Check for missing values
print(df.isnull().sum())
```

```python
# Step 4: Handle missing values
# Drop rows where critical columns are missing (like title/type)
df.dropna(subset=['title', 'type'], inplace=True)

# Fill missing values in less critical columns
df['country'].fillna('Unknown', inplace=True)
df['director'].fillna('Not Specified', inplace=True)
df['cast'].fillna('Not Specified', inplace=True)
```

```python
# Step 5: Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# Step 6: Standardize text values
df['type'] = df['type'].str.strip().str.title()  # e.g., Movie → Movie, tv show → Tv Show
df['country'] = df['country'].str.strip().str.title()  # standardize country names
```

```python
# Step 7: Convert date formats
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
# Step 8: Rename column headers
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# Step 9: Check and fix data types
df['release_year'] = df['release_year'].astype(int)
```

```python
# Step 10: Save the cleaned dataset
df.to_csv("cleaned_netflix_titles.csv", index=False)

print("✅ Cleaning complete! File saved as cleaned_netflix_titles.csv")
```
